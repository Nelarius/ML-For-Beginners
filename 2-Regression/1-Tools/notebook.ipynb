Read scikit-learn documentation about [linear models](https://scikit-learn.org/stable/modules/linear_model.html#linear-model)

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
bunch = datasets.load_diabetes()
print(X.shape)
print(X[0])
```

```python
print(y)
print(bunch["feature_names"])
```

```python
bmi = X[:, np.newaxis, 2]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(bmi, y, test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A graph plot showing diabetes progression against BMI')
plt.show()
```

```python
bp = X[:, np.newaxis, 3]

X_train, X_test, y_train, y_test = model_selection.train_test_split(bp, y, test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

plt.scatter(X_test,y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.title('BP vs disease progression')
plt.show()
```
